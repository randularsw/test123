```python
def train(df,_x,_y,t):
  df = df[[_x,_y]]
  from sklearn.model_selection import train_test_split
  from sklearn.linear_model import LinearRegression
  from sklearn.metrics import mean_squared_error
  X = df[[_x]].values
  Y = df[[_y]].values
  train_features , test_features ,train_labels, test_labels = train_test_split( X , Y , test_size=t )

  X = tf.constant( train_features , dtype=tf.float32 )
  Y = tf.constant( train_labels , dtype=tf.float32 ) 
                                                            
  test_X = tf.constant( test_features , dtype=tf.float32 ) 
  test_Y = tf.constant( test_labels , dtype=tf.float32 )

  reg = LinearRegression(normalize=True).fit(X, Y)
  reg.score(X,Y)
  y_pred = reg.predict(test_X)
  mse = mean_squared_error(test_Y, y_pred)
  predictions = reg.predict(df[[_x]].values)
  return predictions,mse,reg
```

```python
from sklearn import linear_model
```

```python
def train(df,_x,_y,t):
  df = df[[_x,_y]]
  from sklearn.model_selection import train_test_split
  from sklearn.linear_model import LinearRegression
  from sklearn.metrics import mean_squared_error
  clf = linear_model.Lasso(alpha=t)
  X = df[[_x]].values
  Y = df[[_y]].values
  train_features , test_features ,train_labels, test_labels = train_test_split( X , Y , test_size=0.2 )

  X = tf.constant( train_features , dtype=tf.float32 )
  Y = tf.constant( train_labels , dtype=tf.float32 ) 
                                                            
  test_X = tf.constant( test_features , dtype=tf.float32 ) 
  test_Y = tf.constant( test_labels , dtype=tf.float32 )

  reg = clf.fit(X, Y)
  reg.score(X,Y)
  y_pred = reg.predict(test_X)
  mse = mean_squared_error(test_Y, y_pred)
  predictions = reg.predict(df[[_x]].values)
  return predictions,mse,reg
```

```python
X = df1[['VWAP-Stock-1']].values
Y = df1[['Close-Stock-1']].values
train_features , test_features ,train_labels, test_labels = train_test_split( X , Y , test_size=.2 )
vwap_normalizer = preprocessing.Normalization(input_shape=[1,])
vwap_normalizer.adapt(train_features)
```

```python
def build_and_compile_model(norm):
  model = keras.Sequential([
      norm,
      layers.Dense(64, activation='relu'),
      layers.Dense(64, activation='relu'),
      layers.Dense(1)
  ])

  model.compile(loss='mean_absolute_error',
                optimizer=tf.keras.optimizers.Adam(0.001))
  return model
```

```python
model_dnn = build_and_compile_model(vwap_normalizer)
```
